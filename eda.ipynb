````
# Import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

````

````
# Load dataset into a pandas DataFrame
df = pd.read_csv("your_dataset.csv")

````

````
# Check for null values
print(df.isnull().sum())

````

````
# Plot distributions of numeric variables using histograms
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols].hist(bins=20, figsize=(10,10))

````

````
# Plot correlation matrix for all numeric features
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True)

````

````
# Plot distribution of categorical variables using bar plots
categorical_cols = df.select_dtypes(exclude=np.number).columns
for col in categorical_cols:
    sns.countplot(x=col, data=df)
    plt.show()
    
````
